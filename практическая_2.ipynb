```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
```

```python
plt.plot([1,2,3,4,6],[2,4,1,3,2])
plt.xlabel('Ось Х')
plt.ylabel('Ось Y')
plt.title('График массива точек: х = [1,2,3,4,6], y = [2,4,1,3,2]')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2,	17.6,	8.8,	8,	7.7,	6.7]
ax.bar(languages, popularity)

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth= '0.5', color='red')

plt.grid(which='minor', linestyle=':', linewidth= '0.5', color='black')

for language, pop in zip(languages, popularity):
    ax.text(language, pop + 0.3, f'{pop:.1f}', ha='center', va='bottom')

plt.legend()
plt.title("Popularity of Programming Language Worldwide, Oct 2017 compared to a year ago")
plt.xlabel("Languages")
plt.ylabel("Popularity")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2,	17.6,	8.8,	8,	7.7,	6.7]
explode = [0.1 if lang == 'Java' else 0 for lang in languages]

ax.pie(popularity, labels = languages, autopct='%1.2f%%', startangle=140, explode=explode)

plt.show()
```

```python
import random
num_points = 10

group1_mass = [random.randint(55, 75) for _ in range(num_points)]
group1_height = [random.randint(80, 260) for _ in range(num_points)]

group2_mass = [random.randint(55, 75) for _ in range(num_points)]
group2_height = [random.randint(80, 260) for _ in range(num_points)]

group3_mass = [random.randint(55, 75) for _ in range(num_points)]
group3_height = [random.randint(80, 260) for _ in range(num_points)]

plt.scatter(group1_mass, group1_height, marker='*', color='red')
plt.scatter(group2_mass, group2_height, marker='*', color='green')
plt.scatter(group3_mass, group3_height, marker='*', color='blue')

plt.title("Group wise Weight vs Height scatter plot")
plt.xlabel("weight")
plt.ylabel("height")

plt.xticks(range(55, 80, 5))
plt.yticks(range(80, 280, 20))

plt.show()
```
